### Step 1: Generate the County objects

```python
# Step 1: for each county, set the following attributes for each
# (1) population: STATIC
# (2) ncovid: start with the number infected on the START DATE of simulation
# (3) nhealthy: population - ncovid
# (4) county_fip: STATIC
# (5) prob_oc: probability of contracting COVID in a given county on a given DAY

# create Counties from the object County
ca_county_dict = {}
for index, row in init_dat.iterrows():
    county = County('20-Jun-2020', row['county_fips'], row['poc06202020'], row['pop2019'], row['ncovid06202020'])
    ca_county_dict[county.county_fip] = county
```

### Step 2: Generate the agent population

```python
from epigeosim.epigeosim.epiagent import EpiAgent
```

```python
# STEP 2: generate the population for the simplest case:
# - only include the following attributes for each individual:
# (1) origin_fip
# (2) covid_status: once == 1, randomly choose one day in 10, 
#                   and increase the ncovid of the curr_county
# (3) curr_fip
# (4) r_eff (dependent on home county): only populated if covid_status == 1
# (5) days_sick: once == 10, immune_status = 1
# (6) immune_status
# (7) spread_day = randomly chosen day between 1 and 10
# (8) travel_history = a dict of dates and which county they were in

from math import floor

# making the agent population
def gen_pop(df):
    sim_pop = {}
    for index, row in df.iterrows():
        # key for dictionary
        fip = row['county_fips']

        # list of agents for a specific county fip
        fip_pop = []

        ncovid = floor(row['ncovid06202020'])

        print('Beginning fip code: ' + fip)
        if ncovid != 0:
            for j in range(1, ncovid):
                agent = EpiAgent('22-Jun-2020', fip, 1, row['reff06202020'])
                fip_pop.append(agent)
            for k in range(1, row['pop2019']-ncovid):
                agent = EpiAgent('22-Jun-2020', fip, 0)
                fip_pop.append(agent)
        else: 
            for k in range(1, row['pop2019']):
                agent = EpiAgent('22-Jun-2020', fip, 0)
                fip_pop.append(agent)

        print(len(fip_pop))
        sim_pop[fip] = fip_pop
    return sim_pop
```
